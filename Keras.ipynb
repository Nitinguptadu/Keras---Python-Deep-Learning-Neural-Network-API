
# Preprocess data for training with Keras

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_lables = []
train_samples = []
```

```python
for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_lables.append(1)
    
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_lables.append(0)    
    
    



for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_lables.append(1)
    
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_lables.append(0)
    

```

```python
#print(train_samples)
```

```python
#print(train_lables)
```

```python
for i in train_samples:
    print(i)
```

```python
train_lables = np.array(train_lables)
train_samples = np.array(train_samples)
```

```python
#print(train_samples)
```

```python
#print(train_lables)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaler_train_sample = scaler.fit_transform((train_samples).reshape(1,-1))
```

```python
for i in scaler_train_sample :
    print(i)
```

# Create an artificial neural network with Keras

```python
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers import Layer
from keras.layers import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential ([
    Dense(16, input_shape = (1,),activation = 'relu'),
    Dense(32,activation = 'relu'),
    Dense(2,activation = 'softmax')
])
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.00001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(scaler_train_sample,train_lables,validation_split=0.1,batch_size=10,epochs=20,shuffle=True,verbose=2)
```
